```python
ucd = pd.read_csv('used_cars_data.csv')
ucd.info()
```

```python
ucd.isnull()
```

```python
ucd.head()
```

```python
ucd.tail()
```

```python
ucd.isnull().sum()                                           # preprocessing the data
```

```python
(ucd.isnull().sum()/len(ucd)*100)                            # average value
```

```python
ucd.fillna(value = (ucd.isnull().sum()/len(ucd)) * 100, inplace = True)             # the notfilled columns will be filled by (ucd.isnull().sum()/len(ucd)) * 100
```

```python
ucd.isnull().sum()
```

```python
ucd.drop(['S.No.'], axis = 1)                                    # column based operation - axis = 1
```

```python
ucd.drop([0], axis = 0)                                           # row based operation - axis = 0
```

```python
ucd.columns
```

```python
ucd['Fuel_Type']
```

```python
mp.xlabel('Fuel type')
mp.ylabel('Kilometers Driven')
mp.title('Fuel_Type Vs Kilometer_Driven')
mp.bar(ucd['Fuel_Type'], ucd['Kilometers_Driven'], color = 'blue')
```

```python
num_col = ucd.select_dtypes(include = np.number).columns.tolist()
num_col
```
